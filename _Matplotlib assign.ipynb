#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.


##### Matplotlib is a Python library for creating static, animated, and interactive visualizations. It is a popular choice for data visualization in Python, and is used by a wide range of industries, including science, engineering, finance, and healthcare.

#### Matplotlib is used for a variety of reasons, including:

##### To create publication-quality plots
##### To visualize data for analysis and understanding
##### To communicate data to others
##### To create interactive plots that allow users to explore data
##### Five plots that can be plotted using the Pyplot module of Matplotlib include:

##### Line plots: Line plots are used to show the relationship between two variables. The x-axis typically represents the independent variable, and the y-axis represents the dependent variable.
##### Scatter plots: Scatter plots are used to show the relationship between two variables, where the data points are not connected by lines. This can be useful for showing the distribution of data or for identifying clusters of data.
##### Bar plots: Bar plots are used to show the frequency or magnitude of categorical data. The bars are typically stacked, so that the height of each bar represents the total value for a category.
##### Histograms: Histograms are used to show the distribution of numerical data. The x-axis typically represents the data values, and the y-axis represents the number of data points in each bin.
##### Pie charts: Pie charts are used to show the relative proportions of a whole. The slices of the pie represent the different parts of the whole, and the size of each slice is proportional to the value of the part.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
#### import numpy as np
#### np.random.seed(3)
#### x = 3 + np.random.normal(0, 2, 50)
#### y = 3 + np.random.normal(0, 2, len(x))
#### Note: Also add title, xlabel, and ylabel to the plot.


```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
 
plt.scatter(x, y)
 
plt.title("Scatter Plot")
 
plt.xlabel("x label ")
plt.ylabel("y label ")
 
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#### import numpy as np
#### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#### profit = np.array([3000, 8000, 1000, 10000])

```python
import matplotlib.pyplot as plt
import numpy as np
 
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
 
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
 
plt.barh(company, profit)
plt.title("Company Profits")
plt.ylabel("Company")
plt.xlabel("Profit")
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
#### box1 = np.random.normal(100, 10, 200)
#### box2 = np.random.normal(90, 20, 200)

```python
import matplotlib.pyplot as plt
import numpy as np
 
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
 
plt.boxplot([box1, box2])
plt.title("Box Plot")
plt.xlabel("Group")
plt.ylabel("Value")
plt.show()
```
